```python
amz_stock = pd.read_csv('AMZN_stocks_data.csv')
amz_stock.head()
```

```python
# Quiz #i. Create a barplot of the average closing price (Close) for each year. 
```

```python
amz_stock["Year"] = pd.to_datetime(amz_stock["Date"]).dt.year
sns.barplot(data=amz_stock, x='Year', y='Close', estimator='mean')
```

```python
# ii. Create a barplot of the average closing price (Close) for each year, 
# split by whether the high price of the day exceeded $1000$ dollars or not. 
# Hint: Use the hue and estimator parameters of the sns.barplot method.
```

```python
amz_stock["Higher_1000"] = amz_stock["High"] > 1000
sns.barplot(data=amz_stock, x='Year', y='Close', estimator='mean', hue='Higher_1000')
```

```python
# What is the distribution of the `Closing price` (Close) split by # whether the volume of stocks traded exceeded the median volume?
```

```python
sns.boxplot(data = amz_stock,x='Close')
```

```python
median_volume = amz_stock['Volume'].median()
amz_stock['High_volume'] = amz_stock['Volume'] > median_volume
sns.boxplot(data=amz_stock, x="Close", hue="High_volume")
```

```python
sns.boxplot(data=amz_stock, x="High_volume", y="Close")
```

```python
# Tell me the distribution of CLOSE price with bins = 30 # 2 graphs - matplotlib and seaborn!
```

```python
# Create 2 histograms for the opening price: 
# 1) before 2010 
```

```python
avg_by_year = amz_stock.groupby("Year")["Close"].mean().reset_index()
avg_by_year = avg_by_year[avg_by_year["Year"] < 2010]
plt.bar(avg_by_year["Year"], avg_by_year["Close"], color="orange")
```

```python
# 2) incl and after 2010
```

```python
avg_by_year = amz_stock.groupby("Year")["Close"].mean().reset_index()
avg_by_year = avg_by_year[avg_by_year["Year"] >= 2010]
plt.bar(avg_by_year["Year"], avg_by_year["Close"], color="orange")
```

```python
nykaa_data = pd.read_csv('nykaa.csv')
nykaa_data.head()
```

```python
company_counts = nykaa_data['Company Name'].value_counts()
top_10_companies = company_counts.head(10)
top_10_companies
x = nykaa_data['Company Name'].value_counts().values[:10]
x
labels = nykaa_data['Company Name'].value_counts().index[:10]
labels
```

```python
company_names = []
company_values = []

for company, count in top_10_companies.items():
    company_names.append(company)
    company_values.append(count)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(company_values, labels=company_names, autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=90)
plt.title("Top 10 Most Represented Companies in Nykaa")
plt.ylabel('')  # Removes the y-axis label for a cleaner chart
plt.tight_layout()
plt.show()
```
