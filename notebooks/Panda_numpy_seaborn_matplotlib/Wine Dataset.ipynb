```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
reviews = pd.read_csv("wine.csv", index_col=0)
reviews.head()
```

```python
# 1. Bar Chart: Top 10 Wine-Producing Provinces
top_provinces = reviews.groupby('province')['points'].sum().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize = (20, 2))
sns.barplot(data=top_provinces, x='province', y='points')
```

```python
# 2. Bar Chart: Number of Reviews by Score
top_provinces = reviews.groupby('province')['points'].count().sort_values(ascending=False).head(10).reset_index()
top_provinces.rename(columns={'points': 'review_count'}, inplace=True)
plt.figure(figsize = (20, 2))
sns.barplot(data=top_provinces, x='province', y='review_count')
```

```python
# 3. Line Chart: Number of Reviews by Score
```

```python
top_provinces = reviews.groupby('province')['points'].count().sort_values(ascending=False).head(10).reset_index()
top_provinces.rename(columns={'points': 'review_count'}, inplace=True)
plt.figure(figsize = (20, 2))
plt.plot(top_provinces['province'], top_provinces['review_count'], marker='o', color='b', linestyle='-', linewidth=2)
```
